```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
melon_width = np.random.normal(20,0.9,500)
melon_length = np.random.normal(15,0.9, 500)
melon_weight = np.random.normal(3.5, 0.5,500)
```

```python
balloon_width = np.random.normal(18, 2, 500)
balloon_length = np.random.normal(35, 7, 500)
balloon_weight = np.random.normal(0.01, 0.1, 500)
```

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot( projection='3d')
ax.grid()
ax.scatter(  balloon_width , balloon_length ,  balloon_weight ,
            color= 'magenta' , marker="*" , label='balloons',alpha=0.5)
ax.scatter(melon_width , melon_length , melon_weight ,
            color = "lime" , marker="^" , label='melons' , alpha=0.5)
plt.title("balloons and melons")
plt.xlabel("width(cm)")
plt.ylabel("length(cm)")
ax.set_zlabel("weight(kg)", labelpad=-5)

plt.legend() 
plt.show()
```
