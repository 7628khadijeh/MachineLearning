```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("D:/mypy/assigment/assigment38/python.AI/california_cities.csv")
data.head()
```

```python
plt.scatter(data["latd"] ,
              data["longd"] ,
                s=data["area_total_km2"] ,
                  alpha=0.5 ,
                  linewidths=0 ,
                  c=np.log10(data["population_total"]) ,
                  cmap="nipy_spectral" )

plt.colorbar(label="log 10 population")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Area and population of california cities")
plt.show()
```
